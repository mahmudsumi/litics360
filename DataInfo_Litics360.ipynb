# #1: Voter Registration Database (OHIO)
1. Data import
2. Basic info
3. Missing Values

## #1.1: Data import

```python
voterdata_file1 = pd.read_csv("data/SWVF_1_22.txt", sep=",", quotechar='"', header=0, encoding='ISO-8859-1', na_values=['NA'],low_memory=False)
```

```python
voterdata_file2 = pd.read_csv("data/SWVF_23_44.txt", sep=",", quotechar='"', header=0, encoding='ISO-8859-1', na_values=['NA'],low_memory=False)
```

```python
voterdata_file3 = pd.read_csv("data/SWVF_45_66.txt", sep=",", quotechar='"', header=0, encoding='ISO-8859-1', na_values=['NA'],low_memory=False)
```

```python
voterdata_file4 = pd.read_csv("data/SWVF_67_88.txt", sep=",", quotechar='"', header=0, encoding='ISO-8859-1', na_values=['NA'],low_memory=False)
```

```python
voterdata_df = pd.concat([voterdata_file1,voterdata_file2,voterdata_file3,voterdata_file4], axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False, copy=True)
```

## #1.2: Basic info about data

```python
#Check to see if column labeled properly
print('\nColumn labels for Voter Dataset :')
voterdata_df.head()
```

```python
# Quick info about data
print('\nDescription of Voter Dataset :')
voterdata_df.describe()
```

```python
#Number of rows, columns
print('\nVoter Dataset\nTotal # of rows and columns :',voterdata_df.shape)
```

```python
#Data types for each column
print('\nData type information of Voter Dataset :\n')
voterdata_df.info()
```

## #1.3: Basic info about data

```python
#Names of all features
features = list(voterdata_df)

#Missing values for each feature - Output:Bool
na_bool = np.array(voterdata_df.isna().any())

#Missing values for each feature - Output:Number of missing values
na_int = np.array(voterdata_df.isnull().sum())

#Create DataFrame to show missing values
na = list(zip(features,na_bool,na_int))
na_voterdata_df = pd.DataFrame(na, columns = ['feature','na_bool','na_number'])
```

```python
print (na_voterdata_df.to_string())
```
